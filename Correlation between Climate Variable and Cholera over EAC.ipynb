```python
Burundi = d_data.loc["Burundi"]
DRC = d_data.loc["DRC"]
Rwanda = d_data.loc["Rwanda"]
Uganda =d_data.loc["Uganda"]
Kenya=d_data.loc["Kenya"]
Tanzania=d_data.loc[" Tanzania"]
```

```python
# Import library


import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))

# Create figure and multiple plots
fig, axes = plt.subplots(nrows=3, ncols=2)

# Auto adjust
plt.subplot(3, 2, 1)

plt.scatter(Burundi.PCA1 ,Burundi.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("Burundi")




plt.subplot(3, 2, 2)

plt.scatter(Tanzania.PCA1 ,Tanzania.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("Tanzania")



plt.subplot(3, 2, 3)

plt.scatter(Kenya.PCA1 ,Kenya.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("Kenya")



plt.subplot(3, 2, 4)
plt.scatter(Uganda.PCA1 ,Uganda.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("Uganda")

plt.subplot(3, 2, 5)
plt.scatter(Rwanda.PCA1 ,Rwanda.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("Rwanda")


plt.subplot(3, 2, 6)
plt.scatter(DRC.PCA1 ,DRC.NumberOfReportedCasesOfCholera,label='blogs_read', 
         linewidth=3)
plt.title("DRC")


plt.tight_layout()
plt.show()
# Display
```
